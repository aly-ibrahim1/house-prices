# Baseline Model for House Pricing

```python
import pandas as pd
from src.data import load_data

train, test = load_data()
df = pd.concat([train, test])
df.head()
```

```python
from src.features import preprocess, evaluate_model
from sklearn.linear_model import Ridge

df_train = df.loc[train.index, :]
df_test = df.loc[test.index, :]

X_all = df.drop(columns = ['SalePrice']).copy()

X_train = df_train.drop(columns=["SalePrice", "Id"], errors="ignore").copy()
y_train = df_train["SalePrice"].copy()
X_test = df_test.drop(columns=["SalePrice", "Id"], errors="ignore").copy()
y_test = df_test['SalePrice'].copy()

preprocessor = preprocess(X_all)

ridge_model = Ridge(alpha=1.0)
scores = evaluate_model(preprocessor, X_train, y_train, ridge_model)
print(scores)
```
